```python
from qiskit import qiskit,QuantumCircuit
#Create a quantum circuit with 1 qubit and 1 classical bit.
qc=QuantumCircuit(1,1)
#Apply a NOT gate on qubit 0
qc.h(0)
#Measure qubit 0.
qc.measure(0,0)
job=qiskit.execute(qc,qiskit.BasicAer.get_backend('qasm_simulator'))
print(job.result().get_counts())
```

```python
qc.draw()
```

```python
print(qc)
```

```python
#histogram
from qiskit.visualization import plot_histogram
result1=job.result().get_counts()
plot_histogram(result1)
```

```python
result1=job.result().get_counts()
result2=job.result().get_counts()
result3=[result1,result2]
legend=['First execution','Second execution']
title='New histogram'
figsize=(10,10)
color=['crimson','midnightblue']
plot_histogram(result3,legend=legend,title=title,figsize=figsize,color=color)
```
