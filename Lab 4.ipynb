```python
from qiskit import QuantumCircuit,Aer,transpile
qc=QuantumCircuit(2,2)
qc.h(0)
qc.h(1)
```

```python
backend=Aer.get_backend('statevector_simulator')
result=backend.run(transpile(qc,backend)).result()
qcs=result.get_statevector(qc)
```
