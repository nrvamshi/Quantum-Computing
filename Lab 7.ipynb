```python
qc=QuantumCircuit(2,2)
qc.h(0)
qc.x(0)
qc.cx(0,1)
qc.barrier()
qc.measure([0,1],[0,1])
qc.draw()
```

```python
str=qc.qasm()
print(str)
```

```python
qc1=QuantumCircuit.from_qasm_str(str)
```

```python
qc1.draw()
```
